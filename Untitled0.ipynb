```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# スタートするトップレベルのURLを定義
base_url = 'https://www.loft.co.jp/'

response = requests.get(base_url)

# エンコーディングをUTF-8に設定
response.encoding = 'utf-8'

# テキストをHTMLエンティティでデコード
html_decoded = response.text

soup = BeautifulSoup(html_decoded, 'html.parser')

# トップページから全てのリンクを抽出
links = soup.find_all('a', href=True)

# URLを絶対パスに変換し、一意のリストを作成
urls = list(set([urljoin(base_url, link['href']) for link in links]))

# 下層ページの情報を取得
for url in urls:
    # 下層ページにアクセス
    response = requests.get(url)

    # エンコーディングをUTF-8に設定
    response.encoding = 'utf-8'

    page_html = response.text
    page_soup = BeautifulSoup(page_html, 'html.parser')

    # titleタグを抽出
    title = page_soup.find('title').string if page_soup.find('title') else 'No Title Found'

    # 結果を出力
    print(f"URL: {url}")
    print(f"Title: {title}")
    print("-" * 80)
```
